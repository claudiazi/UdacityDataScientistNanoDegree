# Starbucks Capstone Challenge

# Data Sets

The data is contained in three files:

* portfolio.json - containing offer ids and meta data about each offer (duration, type, etc.)
* profile.json - demographic data for each customer
* transcript.json - records for transactions, offers received, offers viewed, and offers completed

Here is the schema and explanation of each variable in the files:

**portfolio.json**
* id (string) - offer id
* offer_type (string) - type of offer ie BOGO, discount, informational
* difficulty (int) - minimum required spend to complete an offer
* reward (int) - reward given for completing an offer
* duration (int) - time for offer to be open, in days
* channels (list of strings)

**profile.json**
* age (int) - age of the customer 
* became_member_on (int) - date when customer created an app account
* gender (str) - gender of the customer (note some entries contain 'O' for other rather than M or F)
* id (str) - customer id
* income (float) - customer's income

**transcript.json**
* event (str) - record description (ie transaction, offer received, offer viewed, etc.)
* person (str) - customer id
* time (int) - time in hours since start of test. The data begins at time t=0
* value - (dict of strings) - either an offer id or transaction amount depending on the record

From above, we can see that if a customer doesnt give their gender, their age will be set as a default number with is 118. Therefore, We want to reset the age with 118 into 'Unknown' and then group the customers into age group (18-25, 26-35, 36-45, 46-55, 56-65, >65, Unknown).

```python
def split_into_age_group(df, bins):
    age_bucket_labels = []
    bins_adj = bins[:-2]
    for i, bin in enumerate(bins_adj):
        if i == 0:
            label = f"18-{bins_adj[i+1]}"
        elif i <= len(bins_adj) - 2:
            label = f"{bin + 1}-{bins_adj[i+1]}"
        elif i == len(bins_adj) - 1:
            label = f"{bin+1}-"
        age_bucket_labels.append(label)
    age_bucket_labels.append("118")
    print(age_bucket_labels)
    df["age_bucket"] = pd.cut(df["age"], bins=bins, labels=age_bucket_labels)
    df["age_bucket"] = df["age_bucket"].replace("118", "Unknown")
    return df
```

```python
bins = [0, 24, 34, 44, 54, 64, 117, 1000]
cleaned_profile = split_into_age_group(profile, bins=bins)
```

#### The average transaction value before a customer received an offer can also be an important feature to influence whether a customer will react to an offer.

```python
transactions = cleaned_transcript[cleaned_transcript["event"]=='transaction']
transactions['value'] = transactions['value'].apply(lambda x: float(x))
transactions.info()
```

```python
# rank the time
transactions = (transactions.assign(rn=transactions.groupby(['person'])['time']
                    .rank(method='first', ascending=True))
                    .sort_values(['person', 'time']))
transactions.head()
```

```python
# get the sum of total transactions
transactions = (transactions.assign(total=transactions.sort_values(['time'])
                                .groupby(['person'])
                                .value
                                .cumsum())
                                .sort_values(['person', 'time']))
transactions['avg_amount'] = transactions['total'] / transactions['rn']
transactions.head()
```

```python
offer_received = cleaned_transcript[cleaned_transcript["event"]=='offer received']
offer_received = offer_received.drop('value_type', axis=1)
offer_received.columns = ['person', 'event', 'offer_id', 'received_time']
offer_received.head()
```

```python
df_merged = pd.merge(offer_received, transactions, on='person')
df_merged = df_merged[(df_merged["received_time"] >= df_merged["time"]) | (df_merged["received_time"]==0)]
```

```python
def plot(df, plt_type, x_col, y_col, plt_title, hue=None, fnt_size=12, fnt_weight='bold'):
    """
    This function creates a bar plot with the mean of a feature for categories
    
    Arguments:
    df: The dataframe with categories
    plt_type: The type of plot (it could be 'lineplot', 'barplot')
    x_col: The name of the column containing the categories
    y_col: The name of the column whose mean should be the y-axis
    plt_title: The plot title
    hue: The name of the column of hue    
    fnt_size: The font size of text (default to 12 point font size)
    fnt_weight: The weight of font (default to 'bold', it could be 'italic')
    
    Returns:
    None
    """
    fig, ax = plt.subplots()    
    y_mean = df.groupby(x_col)[y_col].mean()
    if plt_type == 'lineplot':
        sns.lineplot(data=df, x=x_col, y=y_col, hue=hue)
    if plt_type == 'barplot':
        sns.barplot(data=df, x=x_col, y=y_col, hue=hue)
    plt.title(plt_title)
```

- Define X(input) and Y(ountput)

### Preprocessing for ML  


#### - Hot Encoding 

```python
X["member_year"] = X["member_year"].astype("category")
X["member_month"] = X["member_month"].astype("category")
```

```python
X.info()
```

```python
X = pd.get_dummies(X, prefix_sep="_", drop_first=True)
```

#### - Get train & test set

```python
# split data into train and test sets
from sklearn.model_selection import train_test_split

seed = 42
test_size = 0.3
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

#### - Normalization

```python
from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler().fit(X_train)

# transform training set
X_train_norm = norm.transform(X_train)

# transform test set
X_test_norm = norm.transform(X_test)

# transform the whole set
X_norm = norm.transform(X)
```

### Training the Model

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train RandomForest model
model_rf = RandomForestClassifier(n_estimators = 200, min_samples_leaf=3, random_state=42)
model_rf.fit(X_train_norm, Y_train)
```

```python
# Train XGBClassifier model
from xgboost import XGBClassifier
model_xgb = XGBClassifier(enable_category=True, random_state=42)
model_xgb.fit(X_train_norm, Y_train)
```

```python
# Train XGBClassifier model
from sklearn.tree import DecisionTreeClassifier 
model_dt = DecisionTreeClassifier(random_state = 42)
model_dt.fit(X_train_norm, Y_train)
```

### Model performace evaluation

```python
#baseline
import numpy as np
from sklearn.dummy import DummyClassifier
dummy_clf = DummyClassifier(strategy="stratified", random_state=42)
dummy_clf.fit(X, Y)
dummy_clf.score(X, Y)
```

```python
from sklearn.metrics import confusion_matrix
from tabulate import tabulate
```

```python
def evaluate_models(X, Y, test_X, test_Y, models, models_list):

    """ 
    Description: This function can be used to evaluate a list of classifier models
    Arguments:
    X: features
    Y: classification
    test_X: test set of features
    test_Y: test set of classification
    models: list of models
    Return: 
    a list of accuracy and a list of csv score of 
    """
    accuracy = []
    cvs = []
    precision = []
    recall = []
    f1_score = []
    for i, model in enumerate(models, start=0):
        cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=1)
        n_scores = cross_val_score(model, X, Y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
        Y_pred = model.predict(X_test_norm)
        acc_score = accuracy_score(Y_pred, Y_test)
        accuracy.append(acc_score)
        cvs.append(np.mean(n_scores))
        print(f"Model {i + 1} {models_list[i]}: ")
        print(f"Accuracy: {acc_score:.2f}")
        print(f"Cross validation score: {np.mean(n_scores):.2%} (+/- {np.std(n_scores)*2:.2%})")
        
        cm = confusion_matrix(Y_pred, Y_test)

        if len(cm) == 2:
            prec_score = cm[1, 1] / sum(cm[:, 1])
            recall_score = cm[1, 1] / sum(cm[1, :])
            f1 = 2 * prec_score * recall_score / (prec_score + recall_score)
        print(
            tabulate(
                [
                    ["Accuracy", acc_score],
                    ["Precision", prec_score],
                    ["Recall", recall_score],
                    ["f1 score", f1],
                ],
                headers=["Metric", "Score"],
            )
        )
        precision.append(prec_score)
        recall.append(recall_score)
        f1_score.append(f1)
    return accuracy, cvs, precision, recall, f1_score
```

```python
models_list = ['RandomForest', 'XGBClassifier', 'DecisionTree']
```

```python
accuracy, cvs, precision, recall, f1_score = evaluate_models(X_norm, Y, X_test_norm, Y_test, [model_rf, model_xgb, model_dt], models_list)
```
